### Steps covered:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

sns.set_theme(style='darkgrid')
print('Libraries loaded ')
```

## 1. Load Raw Data

```python
df = pd.read_csv('../dataset/raw_data/Algerian_forest_fires_dataset.csv', skiprows=1)
df.columns = df.columns.str.strip()
print(f'Raw shape: {df.shape}')
df.head()
```

## 2. Remove Header Rows & Strip Whitespace

```python
# Drop region label rows
df = df[~df['day'].str.contains('day|Region', na=True, case=False)]
df = df.dropna().reset_index(drop=True)

# Strip whitespace from strings
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
print(f'After removing bad rows: {df.shape}')
```

## 3. Convert Columns to Numeric

```python
numeric_cols = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws',
                'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna().reset_index(drop=True)
print(f'After conversion: {df.shape}')
df.dtypes
```

## 4. Encode Target Variable

```python
df['Classes'] = df['Classes'].str.strip().str.lower()
df['Classes'] = df['Classes'].map({'fire': 1, 'not fire': 0})

print('Class distribution after encoding:')
print(df['Classes'].value_counts())

# Plot
fig, ax = plt.subplots(figsize=(5, 3))
df['Classes'].value_counts().plot(kind='pie', autopct='%1.1f%%',
    colors=['#2E8B57', '#E05C2A'], ax=ax, labels=['Not Fire', 'Fire'])
ax.set_title('Fire vs Not Fire (%)', fontweight='bold')
plt.tight_layout()
plt.show()
```

## 5. Save Processed Data

```python
df.to_csv('../dataset/processed_data/cleaned_forest_fires.csv', index=False)
print(' Processed data saved to dataset/processed_data/cleaned_forest_fires.csv')
```

## 6. Feature Selection & Train-Test Split

```python
feature_cols = ['Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']
X = df[feature_cols]
y = df['Classes']

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f'Training samples : {X_train.shape[0]}')
print(f'Testing samples  : {X_test.shape[0]}')
print('\n Preprocessing complete. Proceed to visualization.ipynb')
```
