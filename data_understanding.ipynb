```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
print('Libraries loaded ✅')
```

## 1. Load Dataset

```python
# Download dataset
# Source: https://archive.ics.uci.edu/ml/datasets/Algerian+Forest+Fires+Dataset++
# Place the file at: ../dataset/raw_data/Algerian_forest_fires_dataset.csv

df = pd.read_csv('../dataset/raw_data/Algerian_forest_fires_dataset.csv', skiprows=1)
df.columns = df.columns.str.strip()
print(f'Shape: {df.shape}')
df.head(10)
```

## 2. Dataset Info

```python
df.info()
```

```python
# Check missing values
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0] if missing.any() else 'No missing values!')
```

## 3. Statistical Summary

```python
df.describe()
```

## 4. Target Class Distribution

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['Classes'].value_counts().plot(kind='bar', color=['#2E8B57', '#E05C2A'], ax=ax, edgecolor='white')
ax.set_title('Class Distribution: Fire vs Not Fire', fontsize=13, fontweight='bold')
ax.set_xlabel('Class')
ax.set_ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../outputs/graphs/class_distribution.png', dpi=150)
plt.show()
print('\nClass counts:')
print(df['Classes'].value_counts())
```

## 5. Feature Description

| Feature | Description |
|---|---|
| day, month, year | Date of observation |
| Temperature | Noon temperature in °C |
| RH | Relative humidity in % |
| Ws | Wind speed in km/h |
| Rain | Rainfall in mm |
| FFMC | Fine Fuel Moisture Code |
| DMC | Duff Moisture Code |
| DC | Drought Code |
| ISI | Initial Spread Index |
| BUI | Buildup Index |
| FWI | Fire Weather Index |
| Classes | Target (fire / not fire) |
